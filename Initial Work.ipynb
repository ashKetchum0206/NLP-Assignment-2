```python
from nltk.corpus import stopwords
import random
from sklearn.model_selection import train_test_split
random.seed(69)
```

```python
stops = set(stopwords.words('/home/rachit0206/nltk_data/corpora/stopwords/english'))
```

```python
proc_sents = []
for sentence in sents:
    new_sent = ['<s>']
    for token in sentence:
        if(token not in string.punctuation and token not in stops):
            new_sent.append(token)
    
    new_sent.append('</s>')
    proc_sents.append(new_sent)
```

```python
proc_sents_train, proc_sents_valid = train_test_split(proc_sents, test_size=0.2)
```

```python
word_to_int = {}
count = 0
for sentence in proc_sents_train:
    for token in sentence:
        if(token not in word_to_int): 
            word_to_int[token] = count
            count += 1
```

```python
unigrams = {}
threshold = 10
```

```python
for sentence in proc_sents_train:
    for token in sentence:
        if(token not in unigrams): unigrams[token] = 0
        unigrams[token] += 1

to_delete = []
for key in unigrams.keys():
    if(unigrams[key] < threshold):
        to_delete.append(key)
           
```

```python
for token in to_delete:
    unigrams.pop(token)
```

```python
quadgrams = {}
```

```python
for sentence in proc_sents_train:
    for i in range(3 , len(sentence)):
        quadgram = (sentence[i-3] , sentence[i-2] , sentence[i-1], sentence[i])
        if(quadgram not in quadgrams): quadgrams[quadgram] = 0
        quadgrams[quadgram] += 1
```

```python
to_delete = []
threshold = 10
for key in quadgrams.keys():
    if(quadgrams[key] < threshold):
        to_delete.append(key)
        
for quadgram in to_delete:
    quadgrams.pop(quadgram)
```

```python
total_count = 0
for key in quadgrams.keys():
    total_count += quadgrams[key]    
```

```python
import sys
total_perp = 0
count_sent = 0
for sentence in proc_sents_valid:
    
    count_sent += 1
    if(len(sentence) < 4): continue
    else:
        perp = 1
        n = len(sentence)
        for i in range(3 , n):
            n_gram = (sentence[i-3] , sentence[i-2] ,sentence[i-1], sentence[i])
            
            if(n_gram not in quadgrams): 
                perp = 10**7
                break
                
            else:
                perp *= 1/(quadgrams[n_gram]/total_count)
                
        total_perp += perp **(1/n)
        
avg_perp = total_perp / count_sent
```
